```python
# Statistics Basic-2 Assignment

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Q1. What are the three measures of central tendency?

# Answer:
# The three measures of central tendency are:
# 1. Mean
# 2. Median
# 3. Mode

# Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of 
# a dataset?

# Answer:
# Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

## The mean, median, and mode are all measures of central tendency, but they differ in their calculation and interpretation:
'''
1. Mean:
   - Definition: The arithmetic average of all values in a dataset.
   - Calculation: Sum of all values divided by the number of values.
   - Use: Best for symmetrical distributions without extreme outliers.
   - Sensitivity: Affected by all values, including outliers.

2. Median:
   - Definition: The middle value when the dataset is ordered.
   - Calculation: For odd number of values, it's the middle value. For even number, it's the average of the two middle values.
   - Use: Preferred for skewed distributions or when there are extreme outliers.
   - Sensitivity: Not affected by extreme values.

3. Mode:
   - Definition: The most frequently occurring value in the dataset.
   - Calculation: Identify the value that appears most often.
   - Use: Useful for categorical data or discrete numerical data.
   - Sensitivity: Not affected by extreme values.

How they are used:
- Mean is used when we want to take into account every value in the dataset, especially for normally distributed data.
- Median is used when we want to find the central value, particularly useful for skewed data or when dealing with outliers.
- Mode is used to find the most common value, especially useful in categorical data. '''

##Example:
## Let's consider the 
dataset=[1, 2, 2, 3, 4, 5, 5, 5, 6, 100]
import numpy as np
from scipy import stats
## we can use Numpy to calculate mean, median, mode (np.mean(Mean), np.median(Median))
total=0
for i in range(0, len(lis)):
    total=total+lis[i]
print(total)  
'''
Mean=(total/len(dataset))
Median = 4.5 (average of 4 and 5, the two middle numbers when ordered)
Mode = 5 (appears most frequently)'''

## or
print("Mean:----",np.mean(dataset))
print("Median:----", np.median(dataset))
print("mode----", stats.mode(dataset))
'''
In this case:
- The mean is pulled higher by the outlier (100).
- The median gives a better sense of the center for this skewed data.
- The mode shows the most common value. '''



# Q3. Measure the three measures of central tendency for the given height data:

heights = [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

mean = np.mean(heights)
median = np.median(heights)
mode =stats.mode(heights)  # This works for unimodal data

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")

# Q4. Find the standard deviation for the given data:

std_dev = np.std(heights)
print(f"Standard Deviation: {std_dev}")

# Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
 # the spread of a dataset? Provide an example.

# Answer:

import numpy as np

data = [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

range_val = max(data) - min(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Range: {range_val}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")

# These measures describe spread:
# - Range: Total spread of data
# - Variance: Average squared deviation from mean
# - Standard Deviation: Square root of variance, average distance from mean


# Q6. What is a Venn diagram?

# Answer:
"""
A Venn diagram is a visual representation of sets and their relationships. 
It uses overlapping circles or other shapes to show:

1. Elements in each set
2. Common elements between sets (intersections)
3. Elements unique to each set

Key uses:
- Illustrating logical relationships between sets
- Visualizing data in set theory
- Comparing and contrasting groups in various fields

Example: In a class of 30 students, a Venn diagram could show how many play 
football, basketball, both sports, or neither.
"""

# Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:

A = {(2,3,4,5,6,7)}
B = {(0,2,6,8,10)}

# (i) A ∪ B (Union)
union = A.union(B)
print("A ∪ B =", union)

# (ii) A ∩ B (Intersection)
intersection = A.intersection(B)
print("A ∩ B =", intersection)

# Q8. What do you understand about skewness in data?

# Answer:
"""
Skewness in data:
1. Measures asymmetry of the probability distribution
2. Indicates which side of the distribution has a longer tail
3. Types:
   - Positive (right) skew: longer tail on right side
   - Negative (left) skew: longer tail on left side
   - Zero skew: symmetric distribution (e.g., normal distribution)
4. Affects relationship between mean, median, and mode
5. Important for understanding data distribution and choosing appropriate statistical methods
"""

# Q9. If a data is right skewed then what will be the position of median with respect to mean?

# Answer:
"""
In right-skewed data:
- Mean > Median
- The mean is pulled towards the tail (right side)
- The median is typically closer to the peak of the distribution
"""

# Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

# Answer:
"""
Covariance vs Correlation:

Covariance:
- Measures direction of linear relationship between variables
- Unbounded value (-∞ to +∞)
- Affected by scale of variables

Correlation:
- Measures strength and direction of linear relationship
- Bounded value (-1 to +1)
- Scale-independent

Usage:
- Covariance: Understanding if variables move together
- Correlation: Quantifying the strength of relationship between variables

Both used in various statistical analyses, including regression, factor analysis, and portfolio theory.
"""



# Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.

"""
Sample Mean Formula:
x̄ = (Σx) / n

Where:
x̄ = sample mean
Σx = sum of all values in the sample
n = number of values in the sample
"""

# Example calculation
data = [4, 8, 15, 16, 23, 42]

sample_mean = sum(data) / len(data)

print(f"Data: {data}")
print(f"Sample Mean: {sample_mean}")

# Using NumPy for verification
import numpy as np
np_mean = np.mean(data)
print(f"NumPy Mean: {np_mean}")

# Showing the calculation steps
print("\nCalculation steps:")
print(f"Sum of values: {sum(data)}")
print(f"Number of values: {len(data)}")
print(f"Sum / Number of values: {sum(data)} / {len(data)} = {sample_mean}")



# Q12. For a normal distribution data what is the relationship between its measure of central tendency?

"""
In a normal distribution:
- Mean = Median = Mode
- The distribution is symmetric around the central value
"""

# Q13. How is covariance different from correlation?

"""
Covariance vs Correlation:
1. Scale: Covariance is unbounded; Correlation is bounded (-1 to +1)
2. Interpretation: Correlation is easier to interpret (standardized)
3. Units: Covariance is in units of variables; Correlation is unitless
4. Sensitivity: Covariance is sensitive to scale changes; Correlation isn't
"""

# Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

import numpy as np

def demonstrate_outlier_effect(data, outlier):
    print(f"Original data: {data}")
    print(f"Mean: {np.mean(data):.2f}")
    print(f"Median: {np.median(data):.2f}")
    print(f"Standard Deviation: {np.std(data):.2f}")
    
    data_with_outlier = data + [outlier]
    print(f"\nData with outlier: {data_with_outlier}")
    print(f"Mean: {np.mean(data_with_outlier):.2f}")
    print(f"Median: {np.median(data_with_outlier):.2f}")
    print(f"Standard Deviation: {np.std(data_with_outlier):.2f}")

data = [10, 12, 13, 14, 15, 16, 18]
outlier = 100

demonstrate_outlier_effect(data, outlier)

"""
Effects of outliers:
1. Mean: Highly affected, pulled towards the outlier
2. Median: Less affected, robust to outliers
3. Standard Deviation: Increases, showing more spread
4. Mode: Usually unaffected unless the outlier creates a new mode

Choose median and IQR for skewed data or when outliers are present.
"""

```
